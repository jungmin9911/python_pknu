```python
all_numbers = ori[[1, 2, 3, 4, 5, 6, "보너스"]].values.flatten()
all_numbers
```

```python
from collections import Counter
# 등장 횟수 카운트
counter = Counter(all_numbers)

# 1부터 45까지 등장 횟수 정리
result = {num: counter.get(num, 0) for num in range(1, 46)}

# 결과 출력 (숫자별 등장 횟수)
for num, count in result.items():
    print(f"{num:2d}번: {count}회")
```

```python
import json
# JSON 저장
with open("ref/lotto_freq.json", "w", encoding="utf-8") as f:
    json.dump(result, f, indent=2, ensure_ascii=False)
```
